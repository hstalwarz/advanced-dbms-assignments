```python
# import the required libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# dictionary to store the results
result = {'without_index': [119205.3231], 
          'one_index': [108291.0531], 
          'two_indexes': [42622.96993]}
```

```python
# plot the results
plt.style.use('ggplot')

fig, ax = plt.subplots(figsize = (14, 10))

index = np.arange(1)                                          # the x locations of the group
width = 0.3

p1 = ax.bar(index, result['without_index'], 0.2)                   # first set of columns 
p2 = ax.bar(index + width, result['one_index'], 0.2)               # second set of columns
p3 = ax.bar(index + (2 * width), result['two_indexes'], 0.2)       # third set of columns

ax.set_title('Throughput for Bulk Insert Using \n INSERT/SELECT syntax')                         # define the title
ax.set_xticks(index + width)                                         # location of x-ticks
ax.set_xticklabels((''))

ax.legend((p1[0], p2[0], p3[0]), ('Without Index', 'Index on Primary Key', 'Two Indexes'))
plt.xlabel('Indexing Scenarios')                                       
plt.ylabel('Throughput (transactions/second)') 

plt.show()                                                           # show the plot
```
