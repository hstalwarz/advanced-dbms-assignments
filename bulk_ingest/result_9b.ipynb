### 9.b) Multiple Indexes

```python
# dictionary to store the results
result = {'one_index': [6527.757687], 
          'two_indexes': [6451.244359], 
          'three_indexes': [6243.973311],
          'four_indexes': [6426.344743]}
```

```python
# plot the results
plt.style.use('ggplot')


#plt.subplot(2,2,1)
fig, ax = plt.subplots(figsize = (14, 10))

index = np.arange(1)                                          # the x locations of the group
width = 0.2

p1 = ax.bar(index, result['one_index'], width)                   # first set of columns 
p2 = ax.bar(index + (width+0.1), result['two_indexes'], width)               # second set of columns
p3 = ax.bar(index + (2 * (width+0.1)), result['three_indexes'], width)       # third set of columns
p4 = ax.bar(index + (3 * (width+0.1)), result['four_indexes'], width)

ax.set_title('Throughput versus # of indexes')                         # define the title
ax.set_xticks(index + width)                                         # location of x-ticks
ax.set_xticklabels((''))


ax.legend((p1[0], p2[0], p3[0],p4[0]), ('Index on Primary Key', 'Two Indexes', 'Three Indexes', 'Four Indexes'))
plt.xlabel('# of Indexes')                                       
plt.ylabel('Throughput (transactions/second)') 

plt.show()

#plt.subplot(2,2,2)
fig, ax = plt.subplots(figsize = (14, 10))

index = np.arange(1)                                          # the x locations of the group
width = 0.2

p1 = ax.bar(index, result['one_index'], width)                               # first set of columns 
p2 = ax.bar(index + (width+0.1), result['two_indexes'], width)               # second set of columns
p3 = ax.bar(index + (2 * (width+0.1)), result['three_indexes'], width)       # third set of columns
p4 = ax.bar(index + (3 * (width+0.1)), result['four_indexes'], width)        # fourth set of columns

ax.set_title('Throughput versus # of indexes')                         # define the title
ax.set_xticks(index + width)                                         # location of x-ticks
ax.set_xticklabels((''))


ax.set_ylim(6000,7000)


ax.legend((p1[0], p2[0], p3[0],p4[0]), ('Index on Primary Key', 'Two Indexes', 'Three Indexes', 'Four Indexes'))
plt.xlabel('# of Indexes')                                       
plt.ylabel('Throughput (transactions/second)') 
#plt.ylim(top=3)

#plt.ylim(top=3)

plt.show()    
```
